```python
# Load transactions.
transactions = build_sample_transactions('2017-01-01', '2017-12-31')

# Alternatively, upload your own transactions as say 'my_transactions.csv' 
#transactions = read_transactions('my_transactions.csv')

transactions
```

```python
# Summarize over entire period
#
summary = summarize(transactions, decimals=2)
print(summary)
plot(summary)
```

```python
# Summarize again by month start and category
#
summary = summarize(transactions, freq='MS', by_category=True, decimals=2)
print(summary.head(10))
plot(summary)
```

```python
# Find biggest spending categories
#
s = summarize(transactions, by_category=True, decimals=2)

s['debit_frac'] = s['debit']/s['credit'].sum()
s.sort_values('debit_frac', ascending=False)
```
