```python
# Summarize and plot
#
summary = summarize(transactions, freq='MS', by_category=False, decimals=2)
print(summary.head())
plot(summary, currency='NZD')
```

```python
# Summarize again, but also by category
#
summary = summarize(transactions, freq='MS', budget_and_freq=(350, 'W'), 
  by_category=True, decimals=2)
print(summary.head())
plot(summary, currency='NZD')
```

```python
# Find biggest spending categories
#
s = summarize(transactions, by_category=True, decimals=2)

s['debit_frac'] = s['debit']/s['credit'].sum()
s.sort_values('debit_frac', ascending=False)
```
